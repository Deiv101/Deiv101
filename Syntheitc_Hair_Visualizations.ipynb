```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to parse the raw text data into a DataFrame
def parse_raw_data(raw_text):
    # Split the text into lines
    lines = raw_text.strip().split('\n')

    # Find where the column headers start
    header_index = next(i for i, line in enumerate(lines) if 'Response ID' in line)
    headers = lines[header_index].split('\t')

    # Extract data rows (assuming they start with a Response ID)
    data_rows = []
    current_row = {}
    for line in lines[header_index + 1:]:
        if line.startswith('cm') and len(line.split('\t')) > 1:  # New response ID
            if current_row:
                data_rows.append(current_row)
            current_row = {}
            parts = line.split('\t')
            current_row[headers[0]] = parts[0]  # Response ID
            for i, value in enumerate(parts[1:], 1):
                if i < len(headers):
                    current_row[headers[i]] = value
        else:
            # Handle multi-line responses (though rare in this data)
            parts = line.split('\t')
            if len(parts) > 1:
                for i, value in enumerate(parts[1:], 1):
                    if i < len(headers):
                        if headers[i] in current_row:
                            current_row[headers[i]] += ' ' + value
                        else:
                            current_row[headers[i]] = value
    if current_row:
        data_rows.append(current_row)

    # Create DataFrame
    df = pd.DataFrame(data_rows)
    return df

# Your full dataset (pasted here as a string)
raw_data = """
<your entire document text here>
"""  # Replace this with your actual document text

# Parse the data
df = parse_raw_data(raw_data)

# Select only the columns we need for the questions
relevant_columns = {
    '28. Overall, which type of synthetic hair do you prefer?': '28. Overall, which type of synthetic hair do you prefer?',
    '33. What do you typically do with synthetic hair after you no longer use it or need it?': '33. What do you typically do with synthetic hair after you no longer use it or need it?',
    '36. Are you aware of recycling options or initiatives for synthetic hair?': '36. Are you aware of recycling options or initiatives for synthetic hair?',
    '43. Would you support initiatives aimed at improving the sustainability of synthetic hair products?': '43. Would you support initiatives aimed at improving the sustainability of synthetic hair products?',
    '19. Would you consider wearing biodegradable hair for environmental reasons?': '19. Would you consider wearing biodegradable hair for environmental reasons?',
    '45. If you have not used biodegradable wigs or extensions, what is the main reason?': '45. If you have not used biodegradable wigs or extensions, what is the main reason?',
    '40. Would you consider purchasing a second-hand hair extensions (including braids) or wigs?': '40. Would you consider purchasing a second-hand hair extensions (including braids) or wigs?'
}
df = df[list(relevant_columns.values())]

# Analysis functions (same as before, but now using full data)
def plot_preference():
    preference_counts = df['28. Overall, which type of synthetic hair do you prefer?'].value_counts()
    plt.figure(figsize=(8, 6))
    preference_counts.plot(kind='bar')
    plt.title('Overall Synthetic Hair Preference')
    plt.xlabel('Type of Synthetic Hair')
    plt.ylabel('Number of Respondents')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    print("\n1. Synthetic Hair Preference Counts:")
    print(preference_counts)

def plot_disposal():
    disposal_counts = df['33. What do you typically do with synthetic hair after you no longer use it or need it?'].value_counts()
    plt.figure(figsize=(10, 6))
    disposal_counts.plot(kind='bar')
    plt.title('Synthetic Hair Disposal Methods')
    plt.xlabel('Disposal Method')
    plt.ylabel('Number of Respondents')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("\n2. Hair Disposal Counts:")
    print(disposal_counts)

def plot_recycling_awareness():
    awareness_counts = df['36. Are you aware of recycling options or initiatives for synthetic hair?'].value_counts()
    plt.figure(figsize=(8, 6))
    awareness_counts.plot(kind='pie', autopct='%1.1f%%')
    plt.title('Awareness of Recycling Options for Synthetic Hair')
    plt.tight_layout()
    plt.show()
    print("\n3. Recycling Awareness Counts:")
    print(awareness_counts)

def plot_sustainability_support():
    support_counts = df['43. Would you support initiatives aimed at improving the sustainability of synthetic hair products?'].value_counts()
    plt.figure(figsize=(8, 6))
    support_counts.plot(kind='bar')
    plt.title('Support for Sustainability Initiatives')
    plt.xlabel('Response')
    plt.ylabel('Number of Respondents')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    print("\n4. Sustainability Support Counts:")
    print(support_counts)

def plot_biodegradable():
    bio_counts = df['19. Would you consider wearing biodegradable hair for environmental reasons?'].value_counts()
    plt.figure(figsize=(10, 6))
    bio_counts.plot(kind='bar')
    plt.title('Consideration of Biodegradable Hair')
    plt.xlabel('Response')
    plt.ylabel('Number of Respondents')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("\n5. Biodegradable Hair Consideration Counts:")
    print(bio_counts)

def plot_biodegradable_reasons():
    reason_counts = df['45. If you have not used biodegradable wigs or extensions, what is the main reason?'].value_counts()
    plt.figure(figsize=(10, 6))
    reason_counts.plot(kind='bar')
    plt.title('Reasons for Not Using Biodegradable Hair')
    plt.xlabel('Reason')
    plt.ylabel('Number of Respondents')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("\n6. Reasons for Not Using Biodegradable Hair Counts:")
    print(reason_counts)

def plot_second_hand():
    second_hand_counts = df['40. Would you consider purchasing a second-hand hair extensions (including braids) or wigs?'].value_counts()
    plt.figure(figsize=(8, 6))
    second_hand_counts.plot(kind='pie', autopct='%1.1f%%')
    plt.title('Consideration of Second-Hand Hair Products')
    plt.tight_layout()
    plt.show()
    print("\n7. Second-Hand Hair Consideration Counts:")
    print(second_hand_counts)

# Execute all analyses
plot_preference()
plot_disposal()
plot_recycling_awareness()
plot_sustainability_support()
plot_biodegradable()
plot_biodegradable_reasons()
plot_second_hand()

# Print DataFrame info for verification
print("\nDataFrame Info:")
print(df.info())
```

```python
import pandas as pd
import matplotlib.pyplot as plt


residence_column = '5. Place of residence ( city ) '  # Corrected column name

# Function to create pie-of-pie chart for a country
def create_pie_of_pie(country, ax1, ax2):
    sub_df = df[df["Country"] == country]
    sub_counts = sub_df[residence_column].value_counts()
    sub_labels = sub_counts.index
    sub_sizes = sub_counts.values

    # Main pie chart (country)
    ax1.pie([country_counts[country]], labels=[country], autopct='%1.1f%%', startangle=90,
            wedgeprops=dict(width=0.3), pctdistance=0.85, colors=['skyblue'])  # Customize color

    # Secondary pie chart (cities within the country)
    ax2.pie(sub_sizes, labels=sub_labels, autopct='%1.1f%%', startangle=90,
            wedgeprops=dict(width=0.3), pctdistance=0.85, colors=['lightcoral', 'lightgreen', 'gold'])  # Customize colors

    # Draw a connecting line
    plt.plot([0.95, 1.05], [0.5, 0.5], color='gray', lw=1, transform=fig.transFigure, clip_on=False)

# Get country counts
country_counts = df["Country"].value_counts()

# Create subplots for each country
num_countries = len(country_counts)
fig, axes = plt.subplots(num_countries, 2, figsize=(12, 6 * num_countries),
                         gridspec_kw={'width_ratios': [2, 1]})  # Adjust figure size

# Create pie-of-pie charts for each country
for i, country in enumerate(country_counts.index):
    create_pie_of_pie(country, axes[i, 0], axes[i, 1])
    axes[i, 0].set_title(f"Residence in {country}")
    axes[i, 1].set_title(f"Cities in {country}")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest


# 1. Chi-Square Test for Independence (Categorical vs. Categorical)
# Example: Relationship between Education Level and Sustainability Support
# (Columns: '6.  Highest education level achieved ', '43. Would you support initiatives aimed at improving the sustainability of synthetic hair products?')
education_column = '6.  Highest education level achieved '
sustainability_column = '43. Would you support initiatives aimed at improving the sustainability of synthetic hair products?'
contingency_table = pd.crosstab(df[education_column], df[sustainability_column])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print("\n1. Chi-Square Test (Education vs. Sustainability):")
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")

# 2. One-Way ANOVA (Categorical vs. Numerical)
# Example: Differences in Duration of Hair Keeping by Hair Type Preference
# (Columns: '28. Overall, which type of synthetic hair do you prefer?', '27. How long do you typically keep (wigs, extensions, braids, weaves) in your hair before removing or undoing them?')
# Assuming the duration column is numerical (convert if necessary)
hair_type_column = '28. Overall, which type of synthetic hair do you prefer?'
duration_column = '27. How long do you typically keep (wigs, extensions, braids, weaves) in your hair before removing or undoing them?'
groups = df[hair_type_column].unique()
duration_by_type = [df[duration_column][df[hair_type_column] == group] for group in groups]
fvalue, pvalue = stats.f_oneway(*duration_by_type)
print("\n2. One-Way ANOVA (Hair Type vs. Duration):")
print(f"F-statistic: {fvalue}")
print(f"P-value: {pvalue}")

# 3. Two-Sample T-Test (Numerical vs. Numerical with Two Groups)
# Example: Comparing the Number of Wigs Owned by those who are and those who are not aware of recycling options
# (Columns: '36. Are you aware of recycling options or initiatives for synthetic hair?', '17. How many wigs do you own?')
# Assuming 'Number of Wigs Owned' is converted to numerical
recycling_awareness_column = '36. Are you aware of recycling options  or initiatives for synthetic hair?'
wig_ownership_column = '17. How many wigs do you own?'
# (Assuming 'More than...' values in wig ownership are replaced with numerical equivalents)
group1 = df[wig_ownership_column][df[recycling_awareness_column] == 'Yes']
group2 = df[wig_ownership_column][df[recycling_awareness_column] == 'No']
tstatistic, pvalue = stats.ttest_ind(group1, group2)
print("\n3. Two-Sample T-Test (Recycling Awareness vs. Wig Ownership):")
print(f"T-statistic: {tstatistic}")
print(f"P-value: {pvalue}")

# 4. Correlation Analysis (Numerical vs. Numerical)
# Example: Correlation between Age and Willingness to Pay for Sustainability
# (Columns: '4. Age', '42. Would you be willing to pay more for wigs...?')
# Assuming 'Willingness to Pay' is converted to an ordinal scale (e.g., 1-4)
age_column = '4. Age '
willingness_column = '42. Would you be willing to pay more for wigs or extensions that are made from environmentally sustainable materials?'

# Convert 'age_column' to numeric, handling errors
df[age_column] = pd.to_numeric(df[age_column], errors='coerce')

# Convert 'willingness_column' to numeric, handling errors
willingness_mapping = {
    'Yes, definitely': 4,
    'Maybe, depending on other factors (e.g., cost, style, variety)': 3,
    'No, I prefer those that I am used to': 2,
    'Unsure; I need more information': 1
}
df[willingness_column] = df[willingness_column].map(willingness_mapping)


correlation_coefficient, pvalue = stats.pearsonr(df[age_column], df[willingness_column])
print("\n4. Correlation (Age vs. Willingness to Pay):")
print(f"Correlation coefficient: {correlation_coefficient}")
print(f"P-value: {pvalue}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np


# Filter for respondents who have not used biodegradable hair
no_biodegradable = df[df["44. Have you ever used a biodegradable wigs or extensions?"] == " No "]

# Field: "45. If you have not used biodegradable wigs or extensions, what is the main reason?"
# Since this is a multi-select field, split responses, strip spaces, and count occurrences
# Check if no_biodegradable is empty
if no_biodegradable.empty:
    print("No respondents answered 'No' to question 44, therefore there are no reasons to plot.")
else:
    all_reasons = []
    for response in no_biodegradable["45. If you have not used biodegradable wigs or extensions, what is the main reason?"]:
        if isinstance(response, str):
            options = [option.strip() for option in response.split(',')]
            all_reasons.extend(options)

    # Count occurrences of reasons (case-sensitive, space-stripped)
    reasons = pd.Series(all_reasons).value_counts()

    # Create a DataFrame for plotting
    reasons_df = pd.DataFrame({'Reason': reasons.index, 'Count': reasons.values})

    # --- Bar Chart with Color Coding ---
    num_reasons = len(reasons_df)
    colors = cm.get_cmap('viridis', num_reasons)  # Get a colormap with num_reasons colors
    colors = [colors(i) for i in range(num_reasons)]  # Create a list of colors

    plt.figure(figsize=(10, 6))
    plt.bar(reasons_df['Reason'], reasons_df['Count'], color=colors)
    plt.title("Reasons for Not Using Biodegradable Wigs or Extensions")
    plt.xlabel("Reason")
    plt.ylabel("Number of Respondents")
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
    plt.tight_layout()
    plt.show()
```
