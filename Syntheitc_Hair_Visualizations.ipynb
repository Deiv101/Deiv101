```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Data Preparation ---

# City counts from the summary table
city_data = {
    'London': 13, 'Nairobi': 24, 'Birmingham': 2, 'Cardiff': 1, 'Chelmsford': 1, 'Cambridge': 1,
    'Durham': 1, 'Farnborough': 1, 'Uxbridge': 1, 'Wolverhampton': 1, 'Bury St. Edmunds': 1,
    'Dundee': 2, 'Leicester': 1, 'Manchester': 1, 'Nuneaton': 1, 'Swindon': 1, 'Watford': 1,
    'Busia': 1, 'Embu': 3, 'Meru': 1, 'Nakuru': 1, 'Ruiru': 1, 'Thika': 2, 'Chicago': 1,
    'Louisiana': 2, 'Baton Rouge Louisiana USA': 1, 'Accra': 1, 'The Hague': 1, 'Perth': 2, 'Maseru': 1
}

# Mapping cities to countries
country_mapping = {
    'UK': ['London', 'Birmingham', 'Cardiff', 'Chelmsford', 'Cambridge', 'Durham', 'Farnborough',
           'Uxbridge', 'Wolverhampton', 'Bury St. Edmunds', 'Dundee', 'Leicester', 'Manchester',
           'Nuneaton', 'Swindon', 'Watford'],
    'Kenya': ['Nairobi', 'Busia', 'Embu', 'Meru', 'Nakuru', 'Ruiru', 'Thika'],
    'USA': ['Chicago', 'Louisiana', 'Baton Rouge Louisiana USA'],
    'Ghana': ['Accra'],
    'Netherlands': ['The Hague'],
    'Australia': ['Perth'],
    'Lesotho': ['Maseru']
}

# Calculate country totals
country_totals = {country: sum(city_data.get(city, 0) for city in cities)
                  for country, cities in country_mapping.items()}

# Prepare data for double pie chart
outer_labels = list(country_totals.keys())
outer_sizes = list(country_totals.values())
inner_labels = []
inner_sizes = []
for country, cities in country_mapping.items():
    for city in cities:
        if city in city_data:
            inner_labels.append(f"{city} ({country})")
            inner_sizes.append(city_data[city])

# --- Double Pie Chart ---
fig, ax = plt.subplots(figsize=(10, 10))
# Outer pie (countries)
ax.pie(outer_sizes, labels=outer_labels, radius=1.2, wedgeprops=dict(width=0.3, edgecolor='w'),
       autopct='%1.1f%%', pctdistance=1.1, labeldistance=1.2)
# Inner pie (cities)
ax.pie(inner_sizes, labels=inner_labels, radius=0.9, wedgeprops=dict(width=0.3, edgecolor='w'),
       textprops={'fontsize': 4})
plt.title("Double Pie Chart: Country and City Distribution", pad=20)
plt.show()

# --- Additional Visualizations ---

# 1. Bar Chart: Age Distribution
age_data = {'18-25': 38, '26-35': 17, '36-45': 12, '46-55': 3, '56-65': 2}
plt.figure(figsize=(8, 6))
plt.bar(age_data.keys(), age_data.values(), color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()

# 2. Pie Chart: Hair Type
hair_type_data = {
    'Coily, kinky or Afro-textured (Type 4)': 61,
    'Curly (Type 3)': 5,
    'Wavy (Type 2)': 4,
    'Straight (Type 1)': 2
}
plt.figure(figsize=(8, 8))
plt.pie(hair_type_data.values(), labels=hair_type_data.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Hair Type Distribution")
plt.show()

# 3. Stacked Bar Chart: Synthetic Hair Usage by Age Group
synthetic_usage = {
    '18-25': {'Yes': 34, 'No': 4},
    '26-35': {'Yes': 14, 'No': 3},
    '36-45': {'Yes': 11, 'No': 1},
    '46-55': {'Yes': 3, 'No': 0},
    '56-65': {'Yes': 1, 'No': 1}
}
age_groups = list(synthetic_usage.keys())
yes_values = [synthetic_usage[age]['Yes'] for age in age_groups]
no_values = [synthetic_usage[age]['No'] for age in age_groups]

plt.figure(figsize=(10, 6))
plt.bar(age_groups, yes_values, label='Yes', color='lightgreen')
plt.bar(age_groups, no_values, bottom=yes_values, label='No', color='salmon')
plt.title("Synthetic Hair Usage by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.legend()
plt.show()

# 4. Bar Chart: Reasons for Choosing Synthetic Hair
reasons_data = {
    'Protective style': 50, 'Variety of styles/colors': 45, 'Low maintenance': 45,
    'Cost-effective': 30, 'Improve appearance': 30, 'Convenience': 25,
    'Cultural hair': 10, 'Excessive hair loss': 5
}
plt.figure(figsize=(12, 6))
plt.bar(reasons_data.keys(), reasons_data.values(), color='orchid')
plt.title("Reasons for Choosing Synthetic Hair")
plt.xlabel("Reason")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
